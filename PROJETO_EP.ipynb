```python
#Carregamento das bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#Carregamento da base de dados
df = pd.read_csv('content/conjunto_de_dados_de_funcionarios.csv')
df

df.shape

df.info()
```

```python
#Verificar a existência dos valores duplicados e efetuar a remoção

df.duplicated().sum()
```

```python
df[df.duplicated()]

df.drop_duplicates(inplace = True) #Dropping Duplicates
df.info()
```

```python
df.duplicated().sum()
```

```python
#Mudar nome de coluna

df.rename(columns={'Education': 'Formacao'}, inplace=True)
df.info()
```

```python
#Mudar nome de coluna - City

df.rename(columns={'City': 'Cidade'}, inplace=True)
df.info()
```

```python
#Mudar nome de coluna - City

df.rename(columns={'Gender': 'Genero'}, inplace=True)
df.info()
```

```python
#Mudar nome de coluna - City

df.rename(columns={'Age': 'Idade'}, inplace=True)
df.info()
```

```python
df.rename(columns={'JoiningYear': 'Ano de adesão'}, inplace=True)
df.info()
```

```python
df.rename(columns={'PaymentTier': 'Nivel de pagamento'}, inplace=True)
df.info()

df.rename(columns={'LeaveOrNot': 'Sair ou não'}, inplace=True)
df.info()

df.rename(columns={'ExperienceInCurrentDomain': 'Experiencia no Dominio Atual'}, inplace=True)
df.info()
```

```python
#Como agrupar colunas para analisar dados(técnica do Groupby)

cols_=['Formacao','Nivel de pagamento']
filtro_cols=df.filter(items=cols_)
filtro_cols
```

```python
nivel=filtro_cols.groupby(['Formacao'])
nivel.agg({'Nivel de pagamento':['mean','median','std','count']})
```

```python
plt.figure(figsize=(12,6))
nivel.boxplot(subplots=False, rot=45, fontsize=12);
```

```python
anos_idade=df.groupby(['Ano de adesão','Idade'])
anos_idade
```

```python
anos_idade['Nivel de pagamento'].mean()
```

```python
df_1 = df.pivot_table(index='Ano de adesão', columns = 'Idade', values= 'Nivel de pagamento')
plt.figure(figsize=(12,4))
sns.heatmap(df_1,annot=True);

df_1
```
